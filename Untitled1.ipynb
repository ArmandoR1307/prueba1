```python
class coche:
 def __init__(self, marca, modelo, año, color):
  self.marca = marca
  self.modelo = modelo
  self.año = año
  self.color = color

 def descripcion(self):
  print(f"El coche es marca:_ {self.marca}, modelo:_ {self.modelo}, del año:_ {self.año} , y de color:_{self.color} ")

Coche1=coche("Chevrolet", "Tahoe", 2010, "azul")

Coche1.descripcion()
```
